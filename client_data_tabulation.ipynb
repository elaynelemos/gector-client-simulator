```python
import pandas as pd
import numpy as np
```

```python
BASE_DIR = 'outputs/n2d-custom-8-16384'
SCENARIO = 'n2d-custom-8-16384'
NETWORK_LATENCY = 0.02852
```

```python
dfs = []
for i in range(10):
    dfs.append([ pd.read_csv(f'{BASE_DIR}/client-simulation-iter_{i+1}-{2**k}.csv', sep="#") for k in range(10) ])
len(dfs)
```

```python
latency = np.average(dfs[1][1].latency)
latency
```

```python
print(dfs[1][1].latency.to_string(index=False))
```

```python
# latencies = pd.DataFrame(columns=(['sentences']+[*range(1,11)]+['average']))
latencies = pd.DataFrame(columns=(['sentences']+[*range(1,11)]))
latencies.sentences = pd.Series([ 2**i for i in range(10) ])
latencies = latencies.set_index('sentences')
latencies
```

```python
for i in range(10):
    latencies[i+1] = pd.Series([ np.nanmean(dfs[i][k].latency) for k in range(10) ], index=latencies.index)
# latencies.average = latencies.aggregate('mean', axis=1)
latencies
```

```python
latencies = latencies - NETWORK_LATENCY
latencies
```

```python
latencies['average'] = latencies.aggregate('mean', axis=1)
latencies
```

```python
latencies.to_csv(f'{BASE_DIR}/results_client_data-{SCENARIO}.csv', index=True, sep=",")
```
